```python
import pandas as pd 
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('C:\Julio\Python_Intermedio\dataset\Salary_Data.csv')
csv = pd.DataFrame(csv)
print(csv.head())
csv.info()
```

```python
#Tenemos muy pocos valores nulos, los trabajaremos rápido para poder pasar a la parte de analizar
prom_edad = csv['Age'].mean()
prom_años = csv['Years of Experience'].mean()
prom_salary = csv['Salary'].mean()

csv['Age'].fillna(prom_edad, inplace = True )
csv['Years of Experience'].fillna(prom_años, inplace = True)
csv['Salary'].fillna(prom_salary, inplace = True)
```

```python
#Desechamos el resto de datos nulos 
csv.dropna(inplace = True)
print(csv.info())
```

```python
#Comprobamos que ya no tenemos datos nulos
csv.isnull().sum(axis=0)
```
